## repeat for test data 

To recap, we took the following steps to preprocess the training data:
1. remove correlated features 
    - housing_units
    - income_household_six_figure
    - income_household_150_over
    - education_bachelors
    - education_graduate
2. take care of null values
    - drop columns where most of the values are null (threshold = 0.75)
        - metastatic_first_novel_treatment 
        - metastatic_first_novel_treatment_type
    - replace nulls for categorical and numerical data 
3. encode categorical values
    - nominal data --> one-hot encoding

4. standardize the data using StandardScaler
    

To simplify the preprocessing steps, I will create a function called <b>preprocess(DataFrame df)</b>, that will take in the following inputs:
- dataframe to modify
- list of correlated columns to remove 
- null threshold - this will remove the columns whose percentage of null values >= threshold  

The function will return the modified dataframe.
    
Most of the manual work will be left to the function, so you don't have to specify anything besides the list of columns to drop.


```python
# create preprocess function
def preprocess(df, correlated_cols_to_drop, null_cols_to_drop):
    ## drop correlated columns
    df.drop(columns=correlated_cols_to_drop, inplace=True) 
    
    ## drop null cols
    df.drop(columns=null_cols_to_drop, inplace=True)
    
    ## fill in null cols - categorical and numerical
    categorical_cols = df.select_dtypes(include='object').columns.tolist() # select the cols that are categorical (object type)
    numerical_cols = [x for x in df.columns if x not in categorical_cols] # select the rest of the columns in the dataframe (they would be numerical)
    
    # 0 for numerical cols
    for c in numerical_cols:
        # replace numerical NaN with 0
        df[c].fillna(value=0, inplace=True)
    
    # 'None' for categorical cols
    df.replace(np.nan, 'None', regex=True,inplace=True)
    
    ## encode categorical data
    df = pd.get_dummies(df) # this function will one-hot encode the categorical values
    
    ## return the modified dataframe
    return df
```

```python
# test function out on test data - use the correlated_cols_to_drop and null_cols_to_drop variables that we defined earlier.
correlated_cols_to_drop = ['housing_units','income_household_six_figure', 'income_household_150_over','education_bachelors','education_graduate'] # TODO - FILL IN THE COLS WE WANT TO REMOVE
null_cols_to_drop = ['metastatic_first_novel_treatment','metastatic_first_novel_treatment_type']

test_df_final = preprocess(test_df, correlated_cols_to_drop, null_cols_to_drop)
```
