## repeat for test data 

To recap, we took the following steps to preprocess the training data:
1. remove correlated features 
    - housing_units
    - income_household_six_figure
    - income_household_150_over
    - education_bachelors
    - education_graduate
2. take care of null values
    - drop columns where most of the values are null (threshold = 0.75)
        - metastatic_first_novel_treatment 
        - metastatic_first_novel_treatment_type
    - replace nulls for categorical and numerical data 
3. encode categorical values
    - nominal data --> one-hot encoding

4. standardize the data using StandardScaler
    

To simplify the preprocessing steps, I will create a function called <b>preprocess(DataFrame df)</b>, that will take in the following inputs:
- dataframe to modify
- list of correlated columns to remove 
- null threshold - this will remove the columns whose percentage of null values >= threshold  

The function will return the modified dataframe.
    
Most of the manual work will be left to the function, so you don't have to specify anything besides the list of columns to drop.


```python
# create preprocess function
def preprocess(df, correlated_cols_to_drop, null_cols_to_drop):
    ## drop correlated columns
    df.drop(columns=correlated_cols_to_drop, inplace=True) 
    
    ## drop null cols
    df.drop(columns=null_cols_to_drop, inplace=True)
    
    ## fill in null cols - categorical and numerical
    categorical_cols = df.select_dtypes(include='object').columns.tolist() # select the cols that are categorical (object type)
    numerical_cols = [x for x in df.columns if x not in categorical_cols] # select the rest of the columns in the dataframe (they would be numerical)
    
    # median for numerical cols
    for c in numerical_cols:
        df[c].fillna(value=df[c].median(), inplace=True)
    
    # 'None' for categorical cols
    df.replace(np.nan, 'None', regex=True,inplace=True)
    
    ## encode categorical data
    df = pd.get_dummies(df) # this function will one-hot encode the categorical values
    
    ## return the modified dataframe
    return df
```

```python
# test function out on test data - use the correlated_cols_to_drop and null_cols_to_drop variables that we defined earlier.
correlated_cols_to_drop = ['housing_units','income_household_six_figure', 'income_household_150_over','education_bachelors','education_graduate'] # TODO - FILL IN THE COLS WE WANT TO REMOVE
null_cols_to_drop = ['metastatic_first_novel_treatment','metastatic_first_novel_treatment_type']

test_df_final = preprocess(test_df, correlated_cols_to_drop, null_cols_to_drop)
```

## Create function that performs the following:

Step 1 - create the models

Step 2 - fit model to the scaled training data --> X_train_scaled and y_train

Step 3 - use the fitted model to predict the probabilities using the scaled test data --> X_test_scaled 

Step 4 - save the prediction to a file

```python
from datetime import date

# inputs: model_type (string), model (initialized model), array of training data, array of training outputs, array of test inputs, original dataframe of test data    
def create_submission(model_type, model, training_inputs, training_outputs, test_inputs, test_df):
    # fit the model 
    model.fit(training_inputs, training_outputs)
    
    # get the probabilities using the test data
    probabilities = model.predict_proba(X_test_scaled_final)
    print('probabilities: \n', probabilities , '\n')
    
    # the output returned is an matrix showing the probabilities of a 0 or 1. you want to pull the probability of class 1 (aka the diagnosis will be made)
    predictions = probabilities[:,1]
    print('predictions of 1: \n', predictions, '\n')   
    
    # pull only the patient ID column from the original test dataframe. add column for DiagPeriodL90D and set it equal to the predictions
    submission = test_df_final[['patient_id']]
    submission['DiagPeriodL90D'] = predictions
    display(submission)
    
    # save the output into a CSV file. calling my file today's date
    today = date.today()
    print("Today's date:", today)
    
    filename = model_type + '_submission_' + str(today) + '.csv' # name of our output file
    submission.to_csv(filename, index=False) # function to save DF into a csv file. need the index=False parameter to ensure that an extra column of indices does not get added
    print('Submission saved - ' + filename)
    
    # return the trained model
    return model
```

### CatBoostClassifier and RandomizedSearchCV for random param testing

```python
from sklearn.model_selection import RandomizedSearchCV
from catboost import CatBoostClassifier # pip install catboost
from scipy.stats import uniform, randint

# cross validation and parameter tuning example - CatBoost
start_time = time.time() # start timer

catboost_model = CatBoostClassifier()

# Define hyperparameters - this will depend on the model. you can refer to the docs or use ChatGPT to spit out different parameter tests
# these are arbitrary values I got from ChatGPT
param_grid = {
    'iterations': [50, 100, 150, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'depth': [4, 6, 8, 10],
    'l2_leaf_reg': [1, 5, 10],
    'border_count': [5, 10, 20, 50, 100],
    'random_strength': [0.1, 0.5, 1.0],
    'bagging_temperature': [0.2, 0.5, 0.8]
}

# Use RandomizedSearchCV to search for the best hyperparameters - will use n_iter random combinations of  parameters
random_search = RandomizedSearchCV(estimator=catboost_model, param_distributions=param_grid,
                                   scoring='accuracy', n_iter=15, cv=3 )#, random_state=42)
random_search.fit(X_train, y_train)

end_time = time.time() # end the timer
elapsed_time_seconds = end_time - start_time

# print the elapsed time
print(f"\n\nElapsed Time: {int(elapsed_time_seconds // 60)} minutes and {elapsed_time_seconds % 60} seconds\n")

# Print the best parameters and corresponding accuracy
print("Best Parameters: ", random_search.best_params_)
print("Best Accuracy: {:.2f}".format(random_search.best_score_))
```

```python
# create the xgb model using the winning parameters
cb_best_model = random_search.best_estimator_

# call the function- xgb_cv = xgb cross validation
create_submission('catboost_cv', cb_best_model, X_train_scaled_final, y_train, X_test_scaled_final, test_df_final)
```
